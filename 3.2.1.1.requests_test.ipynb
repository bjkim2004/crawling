```python
import requests 
```

```python
r = requests.get('http://hanbit.co.kr') # get() 함수로 웹 페이지를 추출
type(r) # get() 함수의 반환값은 Response 자료
```

```python
r.status_code # status_code 속성으로 HTTP 상태 코드를 확인
```

```python
r.headers['content-type' ] # headers 속성으로 HTTP 헤더를 딕셔 너리로 추출
```

```python
r.encoding # encoding 속성으로 HTTP 헤더를 기반으로 인코딩을 추출
```

```python
r.text # text 속성으로 str 자료형으로 디코딩된 응답 본문을 추출
```
